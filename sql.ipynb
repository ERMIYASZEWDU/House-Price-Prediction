```python
import pandas as pd

# Table A: Customers (Our Left Table)
customers_df = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'CustomerName': ['Ermiyas', 'Almaz', 'Yonas'],
    'ReferrerID': [2, None, 1]  # Almaz referred Ermiyas; Ermiyas referred Yonas
})

# Table B: Orders (Our Right Table)
orders_df = pd.DataFrame({
    'OrderID': [5001, 5002, 5003],
    'CustomerID': [1, 2, 99],    # Note: Customer 99 has no profile; Customer 3 has no orders
    'Item': ['Laptop', 'Smartphone', 'Keyboard'],
    'Price': [1200, 800, 45]
})
```

```python
print("Customers")
print(customers_df)

print("\nOrders")
print(orders_df)
```

## INNER JOIN

```python
pd.merge(customers_df, orders_df, on='CustomerID', how='inner')
```

## LEFT JOIN

```python
pd.merge(customers_df, orders_df, on='CustomerID', how='left')
```

## RIGHT JOIN

```python
pd.merge(customers_df, orders_df, on='CustomerID', how='right')
```

## OUTER JOIN

```python
pd.merge(customers_df, orders_df, on='CustomerID', how='outer')
```

## SELF JOIN

```python
pd.merge(customers_df, customers_df, left_on='ReferrerID', right_on='CustomerID', suffixes=('_NewUser', '_Referrer'))
```

## CROSS JOIN

```python
pd.merge(customers_df, orders_df, how='cross')
```
